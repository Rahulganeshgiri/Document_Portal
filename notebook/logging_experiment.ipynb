```python
import structlog
import os
import datetime
import logging
```

```python
directory = os.path.join(os.getcwd(), 'logs')
directory
```

```python
os.makedirs(directory, exist_ok=True)
```

```python
file_name = f"{datetime.datetime.now().strftime('%m_%d_%Y_%H_%M_%S')}.log"
file_name
```

```python
file_name = os.path.join(directory, file_name)
file_name
```

```python
logging.basicConfig(
    filename=file_name,
    format = "[%(asctime)s] %(levelname)s %(name)s (line: %(lineno)d) - %(message)s",
    level = logging.INFO
)
```
